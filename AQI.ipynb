# forecasting with FB prohet Alogrithm

```python
date_info = pd.to_datetime(air_quality_data['Date'])
print(date_info)
```

```python
time_info = air_quality_data['Time']
```

```python
time_info = time_info.apply(lambda x: x.replace('.',':'))
print(time_info)
```

```python
time_info.info
```

```python
# combining 2 series to a pandas df

date_time = pd.concat([date_info,time_info], axis = 1)
print(date_time)
```

```python
# Combining date & time

date_time['ds'] = date_time['Date'].astype(str)+''+date_time['Time']
date_time.head()
```

```python
data = pd.DataFrame()
```

```python
data['ds'] = pd.to_datetime(date_time['ds'], format='%Y-%m-%d%H:%M:%S')
data.head()
```

```python
data['y'] = air_quality_data['RH']
data.head()
```

```python
from prophet import Prophet
```

```python
model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods = 365, freq = "H")
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
fig1
```

```python
fig2 = model.plot_components(forecast)
fig2
```
